# Gráficos Idade da mãe e Peso do bebê

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

####  1) Base do SINASC do DataSUS, Boxplot da idade da mãe.

```python
dados = pd.read_csv('sinasc_RO_2019.csv')
df = pd.DataFrame(dados)
```

```python
df[['IDADEMAE']].boxplot()
```

####  2) Boxplot do peso do bebê

```python
df[['PESO']].boxplot()
```

####  3) Histograma da idade da mãe

```python
sns.displot(data = df[df['IDADEMAE']>1] 
            ,alpha = .25
            ,x = 'IDADEMAE'
            ,bins = 20
           )
plt.xlabel('Peso da mãe')
plt.ylabel('Quantidade')
```

#### 4) Histograma do peso do bebê

```python

sns.displot(data = df[df['PESO']>1] 
            ,alpha = .25
            ,x = 'PESO'
            ,bins = 30
            ,color='red'
           )
plt.xlabel('Peso (gramas)')
plt.ylabel('Contagem')
```
