# Resposta da primeira solicitação

````python
def obter_dados_da_api(url):
    uf = url.split('=')[1] 
    r = requests.get(url)
    data_json = r.json()
    df = pd.DataFrame(data_json['items'])
    return df
````

````python
obter_dados_da_api('https://apidatalake.tesouro.gov.br/ords/sadipem/tt/pvl?uf=MG&tipo_interessado=Estado')
````

# Resposta da segunda solicitação

````python
df = obter_dados_da_api('https://apidatalake.tesouro.gov.br/ords/sadipem/tt/pvl?uf=MG&tipo_interessado=Estado')
df['status'].value_counts()
````

````python
counts = df['status'].value_counts()
arq_por_dec_prazo = counts['Arquivado por decurso de prazo']
print('Esta registrada', arq_por_dec_prazo, 'solicitação para o Estado de Minas Gerais com status em "Arquivado por decurso de prazo."')
````

# Resposta da terceira solicitação

````python

df = obter_dados_da_api('https://apidatalake.tesouro.gov.br/ords/sadipem/tt/pvl?uf=BA&tipo_interessado=Munic%C3%ADpio')
df
````

````python
df_deferido = df[df['status'] == 'Deferido']
df_cidades_deferido = df_deferido.groupby('interessado').size().reset_index(name='qtd_deferido')

cidade_maior_deferido = df_cidades_deferido.sort_values('qtd_deferido', ascending=False).iloc[0]['interessado']
qtd_maior_deferido = df_cidades_deferido.sort_values('qtd_deferido', ascending=False).iloc[0]['qtd_deferido']

print("A cidade com o maior número de status deferido é:", cidade_maior_deferido)
print("Quantidade de status deferido na cidade:", qtd_maior_deferido)
````
